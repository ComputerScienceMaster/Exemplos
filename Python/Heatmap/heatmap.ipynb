```python
import os
import folium

print(folium.__version__)
```

```python
import pandas as pd

df = pd.read_csv("university.csv", delimiter=",")

df
```

```python
from folium.plugins import HeatMap

m = folium.Map([48., 5.], tiles='stamentoner', zoom_start=6)

HeatMap(df).add_to(m)

m.save(os.path.join('results', 'Heatmap.html'))

m
```
